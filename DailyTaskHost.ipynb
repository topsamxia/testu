```python
import StkDailyTasks as stask
import StkScheduler as ssch
import time, datetime
```

```python
def get_current_combined():
    stask.get_current()
    stask.get_daily_chart_current()
```

```python
stask.get_daily_data()
stask.get_daily_chart()
```

```python
# time data every 2 minutes
ssch.schedule_minute_helper(stask.get_time_current_data, minute_interval='2')
# current data every 10 minutes
ssch.schedule_minute_helper(get_current_combined, minute_interval='6')
# daily task for data
ssch.schedule_daily_helper(stask.get_daily_data, None, '5')
# daily task for chart
ssch.schedule_daily_helper(stask.get_daily_chart, None, '30')

ssch.sched.start()
while(True):
    time.sleep(10000000)
```
