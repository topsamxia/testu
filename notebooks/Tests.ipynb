```python
def getLatestQuote(sk_code):
    s_df = ts.get_realtime_quotes(sk_code)
    s_df = s_df.set_index(['time'])
    return s_df

def readFromData(sk_code, s_filename="", index="time"):
    # make default filename
    if (s_filename==""):
        s_filename = sk_code+".csv"
    return pd.read_csv(s_filename, index_col=index)

def appendData(sk_code, s_filename=""):
    # make default filename
    if (s_filename==""):
        s_filename = sk_code+".csv"
    
    # get latest price
    s_df = getLatestQuote(sk_code)
        
    try:  # read file and then update
        s_dfbase = readFromData(sk_code, s_filename)
        if not(s_df.index[0] in s_dfbase.index):
            s_dfbase = s_dfbase.append(df)
            s_dfbase.to_csv(s_filename)
    except: # 
        s_df.to_csv(s_filename)
```

```python
def processDailyData(tu_inputfile="", defined_exportfile=""):
    
    df = pd.read_csv(tu_inputfile)
    if len(df.index) == 0: pass
    
    def splitDate(x):
        return x.split(" ")[0]
    def splitTime(x):
        return x.split(" ")[1]
    
    series_date = df.date.apply(splitDate)
    series_time = df.date.apply(splitTime)
    
    df["date"] = series_date
    df["time"] = series_time
    
    df.set_index(["date", "time"], inplace = True)
    print (df)
    df.to_csv(defined_exportfile)
    
```
